# Curve fitting with numpy polyfit
## Polyfit is a method that fits the data within a least squares polynomial function.

### Linear

```python
from numpy import linspace, polyfit, poly1d, cos, random
from pylab import plot, show
random.seed(14)
x = linspace( -10 , 10 , 10 )
y = 2*x + 15*random.rand(10)
coeff = polyfit(x,y,1)
xn = linspace(-20,20,100)
yn = poly1d(coeff)
plot(x,y,'o', color = 'cyan')
plot( xn,yn(xn), color = 'red')
```

### Nonleaner qudratic

```python
random.seed(12)
x = linspace( -5 , 5 , 10 )
y = x**2 + 10*random.rand(10)
coeff = polyfit(x,y,2)
xn = linspace(-8,8,100)
yn = poly1d(coeff)
plot(x,y,'o', color = 'magenta')
plot( xn,yn(xn), color = 'blue')
```

### Nonlinear qubic

```python
random.seed(10)
x = linspace( -5 , 5 , 10 )
y = x**3 + 50*random.rand(10)
coeff = polyfit(x,y,3)
xn = linspace(-5,5,100)
yn = poly1d(coeff)
plot(x,y,'o', color = 'red')
plot( xn,yn(xn), color = 'blue')
```
