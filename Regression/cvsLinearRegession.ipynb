```python
from numpy import linspace, polyfit, poly1d, loadtxt, column_stack
from pylab import plot, show, xlabel, ylabel, title
```

```python
fd = loadtxt('./fakedata.csv', delimiter=',')
```

```python
x = fd[:, 0]
y = fd[:, 1]
```

```python
model = polyfit(x,y,1)
xn = linspace(0,15,100)
predict = poly1d(model)
print(predict, '=', 'y')
```

```python
xp = 12
yp = predict(xp)
print(predict, '=', yp)
```

```python
plot(x,y,'o', color = 'cyan')
plot(xn, predict(xn), color = 'red')
plot(xp, predict(xp),'o', color = 'blue')
xlabel("x fake data")
ylabel("y fake data")
title("Fake Data From CVS File")
show()
```
