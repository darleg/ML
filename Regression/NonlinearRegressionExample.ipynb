# Nonlinear Regression Example

### Import functions from the numpy and pylab libraries

```python
from numpy import linspace, polyfit, poly1d, random
from pylab import plot, show
```

### Generate a fake quad dataset for scatter plot

```python
random.seed(12)
x = linspace( 0 , 5 , 10 )
y = x**2 + 10*random.rand(10)
```

### Finding the quad regression line

```python
coeff = polyfit(x,y,2)
xn = linspace(0,8,100)
yn = poly1d(coeff)
print(yn, '=', 'y')
```

### Predict what y would be if x = 7
1. Predict spot on plot is a blue dot

```python
xp = 7
yp = yn(xp)
print(yn, '=', yp)
```

### Plotting

```python
plot(x,y,'o', color = 'magenta')
plot( xn,yn(xn), color = 'green')
plot(xp, yn(xp),'o', color = 'blue')
```

### Generate a fake qubic dataset for scatter plot

```python
random.seed(10)
x = linspace( 0 , 5 , 10 )
y = x**3 + 50*random.rand(10)
```

### Predict what y would be if x = 6

```python
xp = 6
yp = yn(xp)
print(yn, '=', yp)
```

### Plotting

```python
plot(x,y,'o', color = 'red')
plot( xn,yn(xn), color = 'blue')
plot(xp, yn(xp),'o', color = 'magenta')
```
