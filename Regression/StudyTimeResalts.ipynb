# Dataset example
This is a tipical dataset example that is used for showing how machine learning works. It is a small dataset that shows the relationship of student sutdy time to test results. The model is a linear regression and perdicts what a student test results is based on time studding.


## Importing libraries

```python
from numpy import  polyfit, poly1d
from pandas import DataFrame
from pylab import plot, xlabel, ylabel, title
```

##  Dataset
1. Getting the dataset for students study time relations to results.
2. Using pandas DataFrame to setup data.

```python
Student = {'StudyTime': [29, 9, 10, 38, 16, 26, 50, 10, 30, 33, 43, 2, 39, 15, 44, 29, 41, 15, 24, 50],
            'Results': [65, 7, 8, 76, 23, 56, 100, 3, 74, 48, 73, 0, 62, 37, 74, 40, 90, 42, 58, 100]}
StudentData = DataFrame(data=Student)
```

## Defining the feature (x) and target (y) values

```python
x = StudentData.StudyTime
y = StudentData.Results
```

## Machine Learning: Linear Regression

```python
model = polyfit(x, y, 1)
xn = range(0, 51)
predict = poly1d(model)
print(predict, '=', 'y')
```

## Predict what students score would be if student study time is 30 hours
1. Predicted spot on plot is a blue dot

```python
StudentStudyTime = 30
yp = predict(StudentStudyTime)
print(predict, '=', yp)
```

## Plotting

```python

plot(x,y,'o', color = 'cyan')
plot(xn, predict(xn), c = 'red')
plot(StudentStudyTime, predict(StudentStudyTime),'o', color = 'blue')
xlabel("Hours Studding")
ylabel("Test Resualts")
title("Student time studding with test resualts")
```
